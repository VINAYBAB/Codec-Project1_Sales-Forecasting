# Sales Forecasting Project

This notebook predicts future sales using historical time-series data. It incorporates trends, seasonality, and promotional effects.

Models: Prophet (primary) and ARIMA (comparison).

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from prophet import Prophet
from statsmodels.tsa.seasonal import seasonal_decompose
from pmdarima import auto_arima
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load data
df = pd.read_csv('data/sales_data.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
print(df.head())
print(df.info())
```

```python
# Visualize raw data
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['sales'], label='Sales')
plt.scatter(df[df['promotions'] == 1].index, df[df['promotions'] == 1]['sales'], color='red', label='Promotions')
plt.title('Historical Sales with Promotions')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Decompose into trend, seasonal, residual
decomposition = seasonal_decompose(df['sales'], model='additive', period=365)
decomposition.plot()
plt.show()
```

```python
# Split data: Train on first 80%, test on last 20%
train_size = int(len(df) * 0.8)
train = df[:train_size]
test = df[train_size:]
print(f'Train size: {len(train)}, Test size: {len(test)}')
```

```python
# Prophet Model
prophet_train = train.reset_index().rename(columns={'date': 'ds', 'sales': 'y'})
prophet_train['promotions'] = train['promotions'].values

model = Prophet(yearly_seasonality=True, weekly_seasonality=True, daily_seasonality=False)
model.add_regressor('promotions')
model.fit(prophet_train)

# Forecast on test period
future = model.make_future_dataframe(periods=len(test), freq='D')
future['promotions'] = df['promotions'].values  # Use actual promotions for test
forecast = model.predict(future)

# Extract predictions for test
prophet_pred = forecast['yhat'][-len(test):].values

# Evaluate
mae_prophet = mean_absolute_error(test['sales'], prophet_pred)
rmse_prophet = np.sqrt(mean_squared_error(test['sales'], prophet_pred))
print(f'Prophet MAE: {mae_prophet:.2f}, RMSE: {rmse_prophet:.2f}')
```

```python
# ARIMA Model (for comparison)
arima_model = auto_arima(train['sales'], seasonal=True, m=365, trace=False)
arima_pred = arima_model.predict(n_periods=len(test))

# Evaluate
mae_arima = mean_absolute_error(test['sales'], arima_pred)
rmse_arima = np.sqrt(mean_squared_error(test['sales'], arima_pred))
print(f'ARIMA MAE: {mae_arima:.2f}, RMSE: {rmse_arima:.2f}')
```

```python
# Visualize Forecast vs Actual (Prophet)
plt.figure(figsize=(12, 6))
plt.plot(test.index, test['sales'], label='Actual Sales', color='blue')
plt.plot(test.index, prophet_pred, label='Prophet Forecast', color='red')
plt.fill_between(test.index, forecast['yhat_lower'][-len(test):], forecast['yhat_upper'][-len(test):], alpha=0.3, color='red')
plt.title('Prophet Forecast vs Actual Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Plot components
model.plot_components(forecast)
plt.show()
```

```python
# Forecast Future (Next 30 Days)
future_30 = model.make_future_dataframe(periods=30, freq='D')
future_30['promotions'] = 0  # Assume no promotions
forecast_30 = model.predict(future_30)

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['sales'], label='Historical Sales')
plt.plot(forecast_30['ds'], forecast_30['yhat'], label='30-Day Forecast', color='green')
plt.fill_between(forecast_30['ds'], forecast_30['yhat_lower'], forecast_30['yhat_upper'], alpha=0.3, color='green')
plt.title('30-Day Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
